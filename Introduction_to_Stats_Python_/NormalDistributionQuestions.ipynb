```python
np.random.seed(789) 
r = np.random.normal(0, 1, 1000)
```

```python
xseq = np.arange(-4, 4, 0.01)
densities = stats.norm.pdf(xseq, 0, 1)
cumulative = stats.norm.cdf(xseq, 0, 1)
```

```python
ax = sns.lineplot(xseq, densities)
ax.set_title("PDF of Standard Normal")
ax.set_ylabel("Density")
```

```python
ax = sns.lineplot(xseq, cumulative)
ax.set_title("CDF of Standard Normal")
ax.set_ylabel("Cumulative Probability")
```

```python
ax = sns.histplot(r)
ax.set_title("Random draws from Std Normal")
```

```python
## Normal Distribution 
## Q1

mu1 = 45
sigma1 = 10
xseq1 = np.arange(mu1-8*sigma1, mu1+8*sigma1, 0.5)
densities1 = stats.norm.pdf(xseq1, mu1, sigma1)

ax = sns.lineplot(xseq1, densities1)
ax.set_title("PDF of Telephone call Placing")
ax.axvline(mu1)
```

```python
## Q1 a part 

ax = sns.lineplot(xseq1, densities1)
ax.set_title("PDF of Telephone call Placing")
ax.axvline(60, color='red')
```

```python
## Option1 rescale it to Z value and use pnorm 

z1_a = (60 - mu1)/sigma1

stats.norm.cdf(z1_a)
```

```python
## Option 2 directly use cdf

stats.norm.cdf(60, loc = mu1, scale = sigma1)
```

```python
## Q1 b part 

ax = sns.lineplot(xseq1, densities1)
ax.set_title("PDF of Telephone call Placing")
ax.axvline(40, color='red')
```

```python
##Q1 c part 

ax = sns.lineplot(xseq1, densities1)
ax.set_title("PDF of Telephone call Placing")
ax.axvline(70, color='red')
ax.axvline(mu1)
```

```python
## Option1 rescale it to Z value and use cdf 

z1_c = (70 - mu1)/sigma1
1 - stats.norm.cdf(z1_c) # upper tail 
stats.norm.sf(z1_c)
```

```python
z1_c
```

```python
## Option 2 directly use cdf with mean and std

1 - stats.norm.cdf(70, mu1, sigma1)
stats.norm.sf(70, mu1, sigma1)
```

```python
## Question 2

mu2_tr = 5780 + 641 + 712
sigma_tr = sqrt(142**2 + 78**2 + 72**2)
print(mu2_tr)
print(sigma_tr)
xseq2 = np.arange(mu2_tr - 8*sigma_tr, mu2_tr + 8*sigma_tr, 40)
densities2 = stats.norm.pdf(xseq2, mu2_tr, sigma_tr)
ax = sns.lineplot(xseq2, densities2)
ax.set_title("PDF of Total Revenue")
ax.axvline(7000, color="red")
ax.axvline(mu2_tr)
```

```python
z2_c = (7000 -mu2_tr)/sigma_tr
1 - stats.norm.cdf(z2_c)                      # lower tail = F
1 - stats.norm.cdf(7000, mu2_tr, sigma_tr)
```

```python
z2_c = (7000 -mu2_tr)/sigma_tr
stats.norm.sf(z2_c)                      # lower tail = F
stats.norm.sf(7000, mu2_tr, sigma_tr)
```
